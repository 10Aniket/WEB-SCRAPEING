```python
from autoscraper import AutoScraper
```

```python
url = "https://engineering.careers360.com/colleges/list-of-engineering-colleges-in-india"
wanted_list = ["School of Computer Science, University of Petroleum and Energy Studies, Dehradun","Dehradun, Uttarakhand",]
```

```python
scraper = AutoScraper()
```

```python
result = scraper.build(url,wanted_list)
print(result)
```

```python
scraper.get_result_similar("https://engineering.careers360.com/colleges/list-of-engineering-colleges-in-india",grouped=True)
```

```python
scraper.set_rule_aliases({"rule_dy5x" : "colleges_names",  "rule_ow1k": "location"})
scraper.keep_rules(["rule_dy5x","rule_ow1k"])
scraper.save("college web scrapeing")
```

```python
result = scraper.get_result_similar("https://engineering.careers360.com/colleges/list-of-engineering-colleges-in-india",group_by_alias=True)
```
